# Exemplo mais próximo às aplicações práticas da equipe

```python
import networkx as nx
```

```python
DAG = nx.DiGraph([(f"X{x}", "Y") for x in range(1, 7)])
DAG.add_edges_from([('U', 'Y'), ('U', 'T'), ('T', 'Y')])
DAG.add_edges_from([
    ('X1', 'T'),
    ('T', 'X6'),
    ('X2', 'X1'),
    ('X2', 'T'),
    ('Z1', 'X4'),
    ('Z2', 'Z3'),
    ('Z2', 'Z5'),
    ('Z3', 'Z5'),
    ('Z4', 'X6'),
    ('Z5', 'X4'),
    ('Z5', 'Z4'),
    ('Z5', 'X6'),
    ('X5', 'X6')
])
nx.draw(DAG, with_labels=True)
```
